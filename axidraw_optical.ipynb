```python
def make_radiating_lines(A1, A2, n_lines, phase_diff, phase_shift):
    theta = np.linspace(0, 2*np.pi, n_lines) + phase_shift
    x1 = np.cos(theta)*A1
    y1 = np.sin(theta)*A1
    x2 = np.cos(theta + phase_diff)*A2
    y2 = np.sin(theta + phase_diff)*A2
    X = [[_x1,_x2] for _x1,_x2 in zip(x1,x2)]
    Y = [[_y1,_y2] for _y1,_y2 in zip(y1,y2)]
    return X, Y
```

```python
def make_radiating_lines_coord_pairs(A1, A2, n_lines, phase_diff, phase_shift, rev, center):
    theta = np.linspace(0, 2*np.pi, n_lines) + phase_shift
    x1 = (np.cos(theta) * A1 + center[0]).round(4)
    y1 = (np.sin(theta) * A1 + center[1]).round(4)
    x2 = (np.cos(theta + phase_diff) * A2 + center[0]).round(4)
    y2 = (np.sin(theta + phase_diff) * A2 + center[1]).round(4)
    coord_pair_sets = []
    for i in range(len(x1)):
        if rev:
            this_pair_set = [(x2[i],y2[i]),(x1[i],y1[i])]
        elif not rev:
            this_pair_set = [(x1[i],y1[i]),(x2[i],y2[i])]
        coord_pair_sets.append(this_pair_set)
    return coord_pair_sets

def make_radiating_lines_coord_pairs_center_A4(A1, A2, n_lines, phase_diff, phase_shift, rev):
    return make_radiating_lines_coord_pairs(A1, A2, n_lines, phase_diff, phase_shift, rev, (5.5,4.25))
```

```python
# define the initial system state (aka x, y, z positions in space)
initial_state = [0.1, 0, 0]

# define the system parameters sigma, rho, and beta
sigma = 10.
rho   = 28.
beta  = 8./3.

# define the time points to solve for, evenly spaced between the start and end times
start_time = 1
end_time = 60
interval = 100
time_points = np.linspace(start_time, end_time, end_time * interval)
```

```python
# define the lorenz system
def lorenz_system(current_state, t):
    x, y, z = current_state
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return [dx_dt, dy_dt, dz_dt]
```

```python
# return a list in iteratively larger chunks
def get_chunks(full_list, size):
    size = max(1, size)
    chunks = [full_list[0:i] for i in range(1, len(full_list) + 1, size)]
    return chunks
```
