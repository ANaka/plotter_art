```python
class Line():  

    def __init__(
        self,
        start=(0,0),
        end=(0,0)
    ):
        self.x0 = start[0]
        self.y0 = start[1]
        self.x1 = end[0]
        self.y1 = end[1]
        pass
       
    @property
    def xs(self):
        return np.array([self.x0, self.x1])
   
    @property
    def ys(self):
        return np.array([self.y0, self.y1])
   
    @property
    def v0(self):
        return np.array([self.x0, self.y0])
   
    @property
    def v1(self):
        return np.array([self.x1, self.y1])
    
    @property
    def theta_0_1(self):
        return calc_angle(self.v0,self.v1)
    
    @property
    def theta_0_1_deg(self):
        return np.degrees(self.theta_0_1)
```

```python
def calc_angle(v0, v1):
    c = v1 - v0
    return np.arctan2(c[1],c[0])

def calc_midpoint(v0,v1):
    return np.stack([v0,v1]).mean(axis=0)

def calc_distance(v0, v1):
    return np.linalg.norm(v1-v0)
```

```python

def make_circle(r, center, theta_range):
    x_coords = np.cos(theta_range)*r + center[0]
    y_coords = np.sin(theta_range)*r + center[1]
    return np.stack([x_coords, y_coords]).T.round(4)

def make_circle_set(r_range, center, theta_range):
    cp_sets = []
    for r in r_range:
        _cps = make_circle(r, center, theta_range)
        cp_sets.append(_cps)
    return cp_sets
```

```python
def make_radial_line(r_inner, r_outer, center, theta):
    r = np.array([r_inner, r_outer])
    x_coords = np.cos(theta)*r + center[0]
    y_coords = np.sin(theta)*r + center[1]
    return np.stack([x_coords, y_coords]).T.round(4)

def make_radial_line_set(r_inner, r_outer, center, theta_range):
    rl_sets = []
    for theta in theta_range:
        _rl = make_radial_line(r_inner, r_outer, center, theta)
        rl_sets.append(_rl)
    return rl_sets
```

```python
def make_spiral(r_range, center, theta_range):
    x_coords = np.cos(theta_range)*r_range + center[0]
    y_coords = np.sin(theta_range)*r_range + center[1]
    return np.stack([x_coords, y_coords]).T.round(4)
```
